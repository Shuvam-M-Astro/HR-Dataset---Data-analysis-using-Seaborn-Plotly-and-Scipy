```python
# Import necessary modules
import pandas as pd
import seaborn as sns
from scipy import stats
import plotly.express as px

df = pd.read_csv('HRDataset_v14.csv')
```

```python
# Divide dataset based on characteristics
IT = df[df['Department'] == 'IT/IS'] 
Production = df[df['Department'] == 'Production'] 
Software_Engineering = df[df['Department'] == 'Software Engineering'] 
Admin_Offices = df[df['Department'] == 'Admin Offices'] 
Sales = df[df['Department'] == 'Sales'] 
Executive_Office = df[df['Department'] == 'Executive Office'] 
Male = df[df['GenderID'] == 1] 
Female = df[df['GenderID'] == 0] 
```

```python
# Scipy descriptive statistics
stats.describe(df['Salary'].to_numpy())
```
